Area of the House from Basement (in Sqft), Basement Area (in Sqft), Renovated Year , Zipcode  ,Latitude , Longitude may not be  representatives of typical house prices. So I am dropping these columns from dataset df4

```python
x = x.drop(columns=['Area of the House from Basement (in Sqft)','Basement Area (in Sqft)','Renovated Year','Zipcode','Latitude','Longitude'],axis = 1)
```

```python
x.info()
```

```python
plt.boxplot(x)
plt.xticks(rotation=90)
plt.show()
```

```python
def remove_outliers(x,column_name):
  q1=x[column_name].quantile(0.25)
  q3=x[column_name].quantile(0.75)
  iqr=q3-q1
  lower_bound=q1-1.5*iqr
  upper_bound=q3+1.5*iqr
  x[column_name] = x[column_name].clip(upper=upper_bound)
  x[column_name] = x[column_name].clip(lower=lower_bound)
  return x[column_name]
for col in x.columns:
  x[col] = remove_outliers(x,col)

```

```python
plt.boxplot(x)
plt.xticks(rotation=90)
plt.show()
```
